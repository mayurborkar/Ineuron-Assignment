**Question1. Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?.**

Examples

stutter("incredible") ➞ "in... in... incredible?"

stutter("enthusiastic") ➞ "en... en... enthusiastic?"

stutter("outstanding") ➞ "ou... ou... outstanding?"

**Hint :- Assume all input is in lower case and at least two characters long.**


```python
try:
    def stutter(word): #Defining The Function
        """
        These Function Can Be Used For That If Stutters a Word as If Someone Is Struggling to Read It.
        word = word stands for a single word which we can't be pronounced
        """ 
        
        stut = word[0:2] #Slicing The Word
        
        return_string = "{0}... {0}... {1}?".format(stut, word) #Writing The Speicified output
        
        print(return_string)
        print('---'*10)
            
except Exception as e: #Except Block Will Exicuted Only When Their Is Error In Try Block
    print(e)
    
else:  #Thse Will Exicuted When Try Block Exicuted
    print('Their Is No Error In Try Block')
    
finally: #Finally Block Exicuted Whether It Is Error Or Not
    print('Code Is Exicuted Below')
    
#Driver Code

word = 'incredible'
stutter(word)

word = 'enthusiastic'
stutter(word)

word = 'outstanding'
stutter(word)
```

**Question 2.Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place.**

Examples

radians_to_degrees(1) ➞ **57.3**

radians_to_degrees(20) ➞ **1145.9**

radians_to_degrees(50) ➞ **2864.8**


```python
#Importing The Package
import math

try:
    def radians_to_degree(radian): #Defining The Function
        """
        These Function Can Be Used For Convert an Angle In Radians and Returns The Corresponding Angle in Degrees.
        radian = Angle Which We want To Convert From Radian To Degree
        """ 

        π = 3.14159   #Defining The pi Value

        degree = radian * (180/π) #Converting The radian to degree my multiplying With 180/π

        print('radians_to_degrees','(',radian,')','➞',round(degree)) #prinniting The Desired output
        print('---'*10)

except Exception as e: #Except Block Will Exicuted Only When Their Is Error In Try Block
    print(e)
    
else:  #Thse Will Exicuted When Try Block Exicuted
    print('Their Is No Error In Try Block')
    
finally: #Finally Block Exicuted Whether It Is Error Or Not
    print('Code Is Exicuted Below')
    
#Driver Code
radian = 1
radians_to_degree(radian)

radian = 20
radians_to_degree(radian)

radian = 50
radians_to_degree(radian)
```

**Question 3. In this challenge, establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.**

**Given a non-negative integer num, implement a function that returns True if num is a Curzon number, or False otherwise.**

Examples

**is_curzon(5) ➞ True**

2 ** 5 + 1 = 33

2 * 5 + 1 = 11

33 is a multiple of 11

**is_curzon(10) ➞ False**

2 ** 10 + 1 = 1025

2 * 10 + 1 = 21

1025 is not a multiple of 21

**is_curzon(14) ➞ True**

2 ** 14 + 1 = 16385

2 * 14 + 1 = 29

16385 is a multiple of 29


```python
try:
    def is_curzon(num): #Creating The Function
        """
        These Function Can Be Used For Checking given integer num is a Curzon number. The Curzon Number Is Number, 

        If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.

        num = num Stands For Checking The Given Number Is Curzon Number or Not.
        """ 

        power,product = 0, 0 #Defining The Two Variable

        power = 2 ** num + 1 #Calculating The Power Term

        product = 2 * num + 1 #Calculating The Product Term

        if (power % product == 0): #Checking The Condition Whether The Power Is Multiple of Prdoduct or Not
            print('True')
            print(power, 'is a multiple of', product)
            print('---'*10)

        else:
            print('False')
            print(power, 'is not a multiple of', product)
            print('---'*10)
            
except Exception as e: #Except Block Will Exicuted Only When Their Is Error In Try Block
    print(e)
    
else:  #Thse Will Exicuted When Try Block Exicuted
    print('Their Is No Error In Try Block')
    
finally: #Finally Block Exicuted Whether It Is Error Or Not
    print('Code Is Exicuted Below')

#Driver Code
if __name__ == '__main__':
    
    num = 5        #Mention The Number
    is_curzon(num) #Calling The Function
    
    num = 10       #Mention The Number
    is_curzon(num) #Calling The Function
    
    num = 14       #Mention The Number
    is_curzon(num) #Calling The Function
```

**Question 5. Create a function that returns a base-2 (binary) representation of a base-10 (decimal) string number. To convert is simple: ((2) means base-2 and (10) means base-10) 010101001(2) = 1 + 8 + 32 + 128.**

**Going from right to left, the value of the most right bit is 1, now from that every bit to the left will be x2 the value, value of an 8 bit binary numbers are (256, 128, 64, 32, 16, 8, 4, 2, 1).**

Examples

binary(1) ➞ "1"

1 * 1 = 1

binary(5) ➞ "101"

1 * 1 + 1 * 4 = 5

binary(10) ➞ "1010"

1 * 2 + 1 * 8 = 10


```python
class Convert:

    def __init__(self,num):
        self.num = num

    def binary(self):
        try:
            return bin(self.num).replace('0b','')

        except Exception as e:
            print(e)

    def __str__(self):

        return 'The Given Function Is Written Inside The Class & Can Be Used For Converting Binary To Decimal.'
    
#Calling Class
c = Convert(1)
print(c.binary())
print('---'*10)

c = Convert(5)
print(c.binary())
print('---'*10)

c = Convert(10)
print(c.binary())
print('---'*10)
```
