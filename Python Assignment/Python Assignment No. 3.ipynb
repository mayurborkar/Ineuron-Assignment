**2. When does the code in a function run: when it's specified or when it's called?**

**Solution:-** The code in a function run, when the function is called, not when the function is specified.

**Example:-**

```
str1='eat'
str2='ate'
from collections import Counter
def check_anagram(s1,s2): # Here we Just Defined or Specified Function
    s1.lower().replace(" ","") # It is code in Function
    s2.lower().replace(" ","") # It is code in Function
    return Counter(s1)==Counter(s2) 
```

```
check_anagram(str1,str2) # Here We Call The Function
```

**5. In a Python application, how many global scopes are there? How many local scopes are there?**

**Solution:-** Frist we understand that what is global and local scope or variable. 

1] Global Scope:- a) The global scope or variable is noething but an variable that can be created outside of function or class (i.e. created before creating function or class).

b) The global scope can be created only one times in python. The variable is global because any Python function or class defined in this module or notebook, is able to access this variable

**Example:-** 

```
my_name='Mayur' # It Is Global Scope

def my_fun():
  print('My Name Is ',my_name, 'Borkar') 
```

```
my_fun()
```

```
my_name # We can call without calling function also
```

2] Local Scope:- a) The local scope or variable is noething but an variable that can be created inside of function or class (i.e. created after creating function or class).

b) The local scope can be created whenever you call function in python. Variables defined inside a function or class, are not global. Only the function or class can see the variable

**Example:-**

```
def my_second_fun():
  institute='Ineuron' # It Is Local Scope
  print('I Am Student Of',institute,'Institute')
```

```
my_second_fun()
```

```
institute # Here we Can't call local scope outside the function
```

**14. What is the purpose of the try clause? What is the purpose of the except clause?**

**Solution:-** The code that could potentially cause an error goes in the try block & the code that executes if an error happens goes in the except block.

**Example:-**

```
try:
    a=1
    b='m'
    c=a+b
except TypeError:
    print('Try to use both the value with same data type')
except exception as ex:
    print(ex)
 
```
