**1. Given a sentence as txt, return True if any two adjacent words have this property: One word ends with a vowel, while the word immediately after begins with a vowel (a e i o u).**

Examples

vowel_links("a very large appliance") ➞ True

vowel_links("go to edabit") ➞ True

vowel_links("an open fire") ➞ False

vowel_links("a sudden applause") ➞ False

```python
class vowel_links:
    
    def __init__(self, s):
        self.s = s
        
    def vowel_links(self):
        l = self.s.split()
        for i in range(len(l)-1):
            if (l[i][-1] in ['a','e','i','o','u']) and (l[i+1][0] in ['a','e','i','o','u']):
                return True
        else:
            return False
            
v = vowel_links('a very large appliance')
print(v.vowel_links())
print('---'*10)

v = vowel_links('go to edabit')
print(v.vowel_links())
print('---'*10)

v = vowel_links('an open fire')
print(v.vowel_links())
print('---'*10)

v = vowel_links('a sudden applause')
print(v.vowel_links())
print('---'*10)
```

**2. You are given three inputs: a string, one letter, and a second letter. Write a function that returns True if every instance of the first letter occurs before every instance of the second letter.**

Examples

first_before_second("a rabbit jumps joyfully", "a", "j") ➞ True #Every instance of "a" occurs before every instance of "j".

first_before_second("knaves knew about waterfalls", "k", "w") ➞ True

first_before_second("happy birthday", "a", "y") ➞ False #The "a" in "birthday" occurs after the "y" in "happy".

first_before_second("precarious kangaroos", "k", "a") ➞ False

```python
class Frist_Second:
    
    def __init__(self, s, frist_letter, second_letter):
        self.s = s
        self.frist_letter = frist_letter
        self.second_letter = second_letter

    def frist_before_second(self):
        if self.s.rfind(self.frist_letter) < self.s.find(self.second_letter):
            return True
        else:
            return False

v = Frist_Second("a rabbit jumps joyfully", "a", "j")
print(v.frist_before_second())
print('---'*10)

v = Frist_Second("knaves knew about waterfalls", "k", "w")
print(v.frist_before_second())
print('---'*10)

v = Frist_Second("happy birthday", "a", "y")
print(v.frist_before_second())
print('---'*10)

v = Frist_Second("precarious kangaroos", "k", "a")
print(v.frist_before_second())
print('---'*10)
```

**3. Create a function that returns the characters from a list or string r on odd or even positions, depending on the specifier s. The specifier will be "odd" for items on odd positions (1, 3, 5, ...) and "even" for items on even positions (2, 4, 6, ...).**

Examples

char_at_pos([2, 4, 6, 8, 10], "even") ➞ [4, 8] #4 & 8 occupy the 2nd & 4th positions

char_at_pos("EDABIT", "odd") ➞ "EAI" #"E", "A" and "I" occupy the 1st, 3rd and 5th positions

char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd") ➞ ["A", "B", "T", "A", "I", "Y"]

```python
class postion:
    
    def __init__(self,l,s):
        self.l = l
        self.s = s
        
    def char_at_post(self):
        if self.s == 'even':
            return [self.l[i] for i in range(len(self.l)) if (i+1) % 2 ==0]
        else:
            return [self.l[i] for i in range(len(self.l)) if (i+1) % 2 !=0]

p = postion([2,4,6,8,10],'even')
print(p.char_at_post())
print('---'*10)

p = postion('EDABIT','odd')
print(p.char_at_post())
print('---'*10)

p = postion(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"],'odd')
print(p.char_at_post())
print('---'*10)
```

**4. Write a function that returns the greatest common divisor of all list elements. If the greatest common divisor is 1, return 1.**

Examples

GCD([10, 20, 40]) ➞ **10**

GCD([1, 2, 3, 100]) ➞ **1**

GCD([1024, 192, 2048, 512]) ➞ **64**


```python
import math

class Greatest_Common_Divisor:
    
    def __init__(self,lis):
        self.lis = lis
        
    def GCD(self):
        """
        The Given function that returns the greatest common divisor of all list elements. 
        If the greatest common divisor is 1, return 1.
        """
        try:
            max_gcd = 0
            
            for i in range(len(self.lis)-1):
                
                gcd = math.gcd(self.lis[i],self.lis[i+1])
                
                if (gcd > max_gcd):
                    
                    max_gcd = gcd

            return max_gcd
        
        except Exception as e:
            print(e)
            
    def __str__(self):
        
        return 'The Given Function Is Written Inside The Class & Can Be Used For The Finding The GCD.'

# Calling Class
gcd = Greatest_Common_Divisor([10,20,40])
print(gcd.GCD())
print('---'*10)

gcd = Greatest_Common_Divisor([1,2,3,100])
print(gcd.GCD())
print('---'*10)

gcd = Greatest_Common_Divisor([1024, 192, 2048, 512])
print(gcd.GCD())
print('---'*10)
```
